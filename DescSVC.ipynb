```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
plt.style.use('classic')

imagen = Image.open(".\\Imagenes\\zebra.jpg")

plt.imshow(imagen)
```

```python
imagen_gr = imagen.convert('LA')
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)

imagen_mat.shape = (imagen_gr.size[1],imagen_gr.size[0])
```

```python
U, D, V = np.linalg.svd(imagen_mat)
```

```python
print(imagen_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
imagen_reconstruir = np.matrix(U[:,:1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(imagen_reconstruir,cmap='gray')
```

```python
i = 5
imagen_reconstruir = np.matrix(U[:,:i]) * np.diag(D[:i]) * np.matrix(V[:i,:])
plt.imshow(imagen_reconstruir,cmap='gray')
titulo = 'valores singulares = %s'%i
plt.title(titulo)
plt.show()
```
